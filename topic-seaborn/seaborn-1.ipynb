```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('dowjones')
df = df.head(50)
df.head()
```

```python
df = pd.read_csv('insurance.csv').sample(200, random_state=1)
df['children'] = df['children'].astype('str')
df.head(3)
```

```python
"""Figure with one Axes"""
fig, axes = plt.subplots(figsize=(8,8))

sns.scatterplot(data=df, x='age', y='charges', hue='smoker')   # Seaborn code to draw a scatter plot
plt.title("Seaborn Plot!!!")
plt.xlabel('X-Axis: age ')
plt.legend(loc='upper left', title='Legend', frameon=False)
plt.show()
```

```python
"""Figure with two Axes"""
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

sns.scatterplot(data=df,
                x='age',
                y='charges',
                hue="smoker",
                ax=axes[0]) # you use the Axes from Matplotlib line of code above

axes[0].set_title('Nice Title')
sns.histplot(data=df, x="children", ax=axes[1])
sns.histplot(data=df, x="charges", ax=axes[2])

plt.tight_layout()
fig.suptitle('Super title for the Figure', fontsize=16, y=1.1)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('insurance.csv')
df = df.sample(n=100, random_state=6)
df['children'] = df['children'].astype('str')
df.head(3)
```

```python
### Histogram ###

"""The arguments are data, where you parse your dataset, and x, where you parse the variable you are interested in drawing."""

sns.set_theme(style="whitegrid")
sns.histplot(data=df, x='charges')
plt.show()
```

```python
"""If you want to have more than one variable in your plot, you can subset the variables and parse at data argument

    * In this case, we used bracket notation to subset ['age','bmi'].

"""

sns.set_theme(style="whitegrid")
sns.histplot(data=df[['age','bmi']])
plt.show()
```

```python
sns.histplot(data=df, kde=True, x='charges') # kde=True adds a kernel density estimate
plt.show()
```

```python
"""You can understand the distribution of a given category by adding the argument hue

    * In this case, we want to know charges distribution per smoking status.
    * The most frequent levels from non-smokers look different from the smokers. We notice the distribution shape
    of non-smokers is tighter and centred at a lower medical insurance charge.

    * The distribution of smokers is more spread out and has a higher peak at higher charges."""

sns.histplot(data=df, kde=True, x='charges', hue='smoker') # hue='smoker' adds a color distinction based on the 'smoker' column
plt.show()
```

```python
### Displot ###

"""Distplot visualizes the distribution of a single variable by combining a histogram (frequency of intervals) with a KDE (smooth distribution estimate). It's useful for quickly assessing data distribution or comparing it to a normal curve, and can facet plots by subsets for deeper analysis."""

sns.displot(df, x="charges", col="smoker", row="sex", hue='region')
# col="smoker" creates separate columns for each smoking status
# row="sex" creates separate rows based on sex
# hue='region' adds color distinction based on the 'region' column
plt.show()
```

```python
### KDEplot ###

"""kdeplot displays the distribution of a numeric variable as a smooth probability density curve, making it easier to interpret than a histogram. It's ideal for viewing distribution shape, comparing multiple datasets, or examining changes, such as investment returns before and after an intervention."""

df = sns.load_dataset('tips')
df = df.sample(n=50, random_state=1)
df.head(3)

fig,ax = plt.subplots(figsize =(8,6))
sns.kdeplot(data=df.filter(['tip','total_bill'], axis=1),fill=True)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize =(11,6))

sns.kdeplot(data=df, x='tip',hue='sex',ax=axes[0,0],fill=True,  palette='Set1')
sns.kdeplot(data=df, x='tip',hue='time',ax=axes[0,1],fill=True,  palette='Set1')
sns.kdeplot(data=df, x='tip',hue='day',ax=axes[1,0],fill=True,  palette='Set1')
sns.kdeplot(data=df, x='tip',hue='size',ax=axes[1,1],fill=True, palette='Set1')

plt.tight_layout()
plt.show()
```

```python
### Boxplot ###

df = sns.load_dataset('tips')
df.head()
```

```python
sns.boxplot(data=df, x='size', y='tip') # x = the categorical variable to inspect the levels across; y = the numerical variable that we want to assess the distribution
plt.show()
```

```python
sns.boxplot(data=df, x="size", y="tip", hue="time") # drawing a grouped boxplot with hue (a categorical data)
plt.show()
```

```python
df_practice = pd.read_csv('insurance.csv')
df_practice = df_practice.sample(n=50, random_state=1)
df_practice['children'] = df_practice['children'].astype('str')
df_practice.head(3)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(11,6))
sns.boxplot(data=df_practice, x='region', y='charges', hue='smoker', ax=axes[0], palette='Set1')
sns.boxplot(data=df_practice, x='region', y='charges', hue='sex', ax=axes[1], palette='Set1')

plt.tight_layout()
plt.show()
```

```python
### Swarmplot ###

"""A swarmplot is a scatterplot for categorical variables that shows individual data points, providing a clearer view of value distribution. It's useful alongside boxplots to display all observations but isn't ideal for very large datasets.
* The arguments are data, x and y: you can interpret them similarly to the boxplot."""

sns.swarmplot(data=df, x='size', y='tip')
plt.show()
```

```python
sns.swarmplot(data=df, x='size', y='tip', hue='time', dodge=True)
# hue parameter - to draw a grouped swarmplot with the data being categorical
# dodge=True - so dots will not be cluttered 
plt.show()
```

```python
"""You can combine a boxplot and a swarmplot in your analysis. The benefit you take from a boxplot is to assess the distribution levels. The benefit you get from a swarmplot is to understand how frequent the levels are."""

x, y, hue = 'size', 'tip', 'time'
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,6))

sns.boxplot(data=df, x=x, y=y, hue=hue, ax=axes[0])
sns.swarmplot(data=df, x=x, y=y, hue=hue, dodge=True, ax=axes[1])
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_practice = pd.read_csv('insurance.csv')
df_practice = df_practice.sample(n=50, random_state=1)
df_practice = df_practice.drop(columns=['Unnamed: 0'])
df_practice.head(3)
```

```python
### Jointplot ###

"""jointplot combines a scatter plot (or similar) with histograms or KDE plots of each variable, letting you examine both the relationship between two continuous variables and their individual distributions in one visualization."""

"""Example Application: Exploring the relationship between variables like height and weight, or price and quality ratings, where understanding both the correlation and the distribution of each variable independently is important."""

x, y, hue = 'age', 'charges', 'smoker'
sns.set_theme(style='whitegrid')

sns.jointplot(data=df_practice, x=x, y=y, hue=hue)

plt.show()
```

```python
### LM Plot ###

"""lmplot pairs scatter plots with regression lines to visualize linear relationships, and can show how these relationships differ across categories."""

"""Example Application: Useful in business analytics for regression analysis like sales over time by different market segments, or in healthcare for plotting trends of medical measurements across different patient groups."""

df_practice = sns.load_dataset('tips')
df_practice = df_practice.sample(n=50, random_state=1)
df_practice.head(3)
```

```python
x, y, hue = 'total_bill', 'tip', 'sex'

sns.lmplot(data=df_practice, x=x, y=y, hue=hue)

plt.show()
```

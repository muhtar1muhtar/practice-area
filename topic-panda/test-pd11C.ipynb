```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

```python
df_practice = sns.load_dataset('tips')
print(f"DataFrame shape: {df_practice.shape}")
df_practice.head(10)
```

```python
def CalculateTipPerception(df_practice):
    df_practice['TipPerception'] = df_practice['tip'].apply(lambda x: 'Nice' if x > 3 
                                                            else 'Could be better')
    return df_practice
```

```python
df_practice.pipe(CalculateTipPerception)
df_practice.head(10)
```

```python
def my_price_perception(price, cut):
    if cut in ['Premium', 'Ideal'] and price < 2000:
        return "Worth it, buy now"
    elif price < 1000:
        return "Buy"
    elif price < 2500:
        return "Think twice"
    else:
        return "No thanks"
    
df['MyValuePerception'] = df.apply(lambda x: my_price_perception(x['price'], x['cut']), axis=1)
df
```

```python
df['MyValuePerception'].value_counts()
```

```python
df_practice = sns.load_dataset('tips')
print(f"DataFrame shape: {df_practice.shape}")
df_practice.head()
```

```python
def day_value(day):
    if day in ['Sat', 'Sun']:
        return True
    else:
        return False

df_practice['weekend'] = df_practice.apply(lambda x: day_value(x['day']), axis=1)
df_practice.head()
```

```python
df_practice['weekend'].value_counts()
```

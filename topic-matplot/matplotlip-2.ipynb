```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_practice = sns.load_dataset('tips')
df_practice = df_practice.sample(n=50, random_state=1)
df_practice.head(10)
```

```python
day_frequencies = df_practice.value_counts('day', normalize=True)
plt.pie(x=day_frequencies, labels=day_frequencies.index)
plt.show()
```

```python
size_frequency = df_practice['size'].value_counts()
plt.barh(y=size_frequency.index, width=size_frequency, color='purple')
plt.show()
```

```python
fig = plt.figure(figsize=(8,7))
ax = plt.axes(projection='3d')

x = df_practice['total_bill']
y = df_practice['tip']
z = df_practice['size']
ax.scatter(x, y, z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

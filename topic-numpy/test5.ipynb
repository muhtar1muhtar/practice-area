```python
import numpy as np
```

```python
np.random.seed(seed=1)
np.random.randint(low=10,high=50,size=(2,3,4,5)) # Generates a 4D array with shape (2, 3, 4, 5) filled with random integers between 10 and 50
```

```python
np.random.seed(seed=1)
np.random.randint(low=10,high=50,size=(2,2,3,4,5)) # Generates a 5D array with shape (2, 2, 3, 4, 5) filled with random integers between 10 and 50
```
